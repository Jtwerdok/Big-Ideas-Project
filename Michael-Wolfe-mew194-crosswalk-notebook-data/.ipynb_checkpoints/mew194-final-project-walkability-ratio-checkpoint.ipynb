### Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

### Data Transformations

```python
ratios = pd.read_csv("sidewalkstreetratiotract.csv")
#drop NaNs and duplicates
ratios = ratios.dropna()
rations = ratios.drop_duplicates()
#sort from highest to lowest ratio
ratios = ratios.sort_values(by=['Ratio'], ascending=False)

#https://transition.fcc.gov/form477/Geo/more_about_census_tracts.pdf

#conversion to String
ratios["Tract"] = ratios["Tract"].astype(str)
#extraction of local census tract
ratios["Tract"] = ratios["Tract"].str[5:9]
ratios.head(20)

```

```python
#read in data mapping neighborhoods to census numbers
neighborhoods = pd.read_csv("index_pittsburghneighborhoods_blocks_2020.csv")
neighborhoods["TRACT"] = neighborhoods["TRACT"].astype(str)
neighborhoods = neighborhoods.rename(columns = {"TRACT":"Tract"})

#isolated tract number and neighborhoods in a dataframe
neighborhoods = neighborhoods[["Tract", "Neighborhood"]]

#dropped duplicates
neighborhoods = neighborhoods.drop_duplicates()

#appended 0 to beginning of census tracts that were of length 5
neighborhoods['Tract'] = np.where(neighborhoods['Tract'].str.len() == 5, "0" + neighborhoods['Tract'], neighborhoods['Tract'])
neighborhoods['Tract'] = neighborhoods['Tract'].str[0:4]

#merged on Tract column
merged = pd.merge(ratios, neighborhoods, on='Tract')
merged = merged.sort_values(by=['Ratio'], ascending=False)
merged.head(20)
```

### Matplotlib Bar Graph

```python
merged = merged.head(5)
merged = merged.sort_values(by=['Ratio'], ascending=True)
merged['Ratio'] = merged['Ratio'].astype(float)
neighborhoods = list(merged['Neighborhood'])
ratios = list(merged['Ratio'])
  
fig = plt.figure(figsize = (10, 5))
 
plt.bar(neighborhoods, ratios, color ='maroon',
        width = 0.4)
 
plt.xlabel("Neighborhoods")
plt.ylabel("Sidewalk-To-Street Ratio")
plt.title("Top 5 Sidewalk-To-Street Ratios")
plt.autoscale(enable=True, axis='y')

for i in range(len(neighborhoods)):
        plt.text(i,ratios[i],ratios[i])

plt.show()
```
